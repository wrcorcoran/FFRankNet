```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd

from model import RN, get_ranking, save_ranking, make_pairs, train
```

```python
def run_analysis(loc):
  X = np.load(f'../../assets/{loc}/X.npy')
  y = np.load(f'../../assets/{loc}/y.npy')
  mp = pd.read_csv(f'../../assets/{loc}/map.csv')
  final = np.load(f'../../assets/{loc}/finalX.npy')

  xi, xj, labels, id = make_pairs(X, y)
  xi_train, xi_test, xj_train, xj_test, labels_train, labels_test, id_train, id_test = train_test_split(xi, xj, labels, id, test_size=0.2)
  model = train(xi_train, xj_train, labels_train, 200, 32, xi_test, xj_test, labels_test, 25)
  indices = get_ranking(final, model)
  save_ranking(indices, loc, mp)
  del model
```
